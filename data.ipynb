```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

#### Importing the required libraries for the operations on the dataset. numpy is numerical python and helps to perform numerical operations on the dataset. Pandas helps us to work on the dataframes. Matplotlib is used for visualizations of the datasets.

```python
stu=pd.read_csv(r"performance.csv")
type(stu)
```

```python
print(stu.dtypes)
```

#### type keyword in python gives the type of the object and in our case it is a Dataframe object of pandas. With the object if the dtypes keyword is used, it gives the data type of all the columns in the dataset. Reading of the file is done by using the read_csv command in pandas. The 'r' in the reading of file signifies that the file should be read as rawdata.

```python
obj=stu[["math_score","reading_score","writing_score"]]
print("Mean of scores is ")
print(obj.mean())
print()
print("Minimum of scores is ")
print(obj.min())
print()
print("Maximum of scores is ")
print(obj.max())
print()
print("Standard deviation of scores is ")
print(obj.std())
```

#### The operations that can be performed on the dataset are many, Some basic opeartions include mean, Minimum, Maximum and standard  deviation. The operations are performed by using the dataframe object and the column on which the operation needs to be performed.

```python
stu['total_scores']=stu['math_score']+stu['reading_score']+stu['writing_score']
stu.head(5)
```

```python
stu['percentages']=stu['total_scores']/3
stu.tail(5)
```

#### Here we are adding 2 new columns called as total_scores and percentages. The values for these columns are taken from the pre-existing columns of the dataset. The total_scores are calculated by adding the 3 columns Reading, Math and Writing Scores. The percentages are calculated using the total_scores column. The command dataframe object.head(integer) helps us to list only the desired number of rows from the top of dataset. Similarly object.tail(Integer) helps to list the desired number of rows from bottom of dataset.

```python
pd.crosstab([stu.math_score,stu.reading_score,stu.writing_score],stu.gender)
```

```python
pd.crosstab([stu.group,stu.gender],[stu.math_score,stu.reading_score,stu.writing_score])
```

#### The crosstab keyword allows the user to create a view of the columns as a cross tab. The desired columns are arranged in a row and column arrangement and the comparison of the same is showcased. The columns incuded in the crosstab can be an array of columns from the dataset or individual columns based on the requirement.

```python
stu.groupby("group").size()
```

The size function finds the count in the dataset and displays it with respect to the data in the columns. The describe function helps us to find the overall description of the dataset but is grouped based on user's desire. The description gives all kinds of data that are used for the plotting. The 25, 50, 75 percentages are used for the box-plot type of graph.

```python
df=stu.sort_values(['percentages'])
df.head(10)
```

```python
stu.corr()
```

#### corr() function helps to find the corelation between all the columns of the data set. The Corelation will give the normalized values in range of 0-1 and display it as a table. The corr function also helps in plotting a graph.

#### The above code segment counts the number of entries for that particular column of the dataframe object. The Value_counts function also sorts the values in descending order by default. #### The object.to_csv(location) command helps us to store our work as a desired file format. This is helpful if any extra columns are added to the dataset. The ouput of this executable will create a csv file with the required columns. 

```python
size=np.array(stu.groupby("group").size())
labels="Group A","Group B","Group C","Group D","Group E"
colors=["red","green","purple","yellow","grey"]
explode=(0.05,0.05,0.05,0.05,0.05)
plt.pie(size,explode=explode,labels=labels,colors=colors,shadow="True",autopct="%.f%%")
plt.show()
```

#### The pie chart is plotted for the group and the number of students in each group. The numpy array command is used to create an array of values for the count of the students. Here both the pandas dataframe and the numpy array are used together to get a combined result. The result is a pie chart of the groups and their number of students in percentages. 

```python
plt.hist(stu.math_score,color='blue')
plt.hist(stu.reading_score,color='orange')
plt.hist(stu.writing_score,color='red',bins=20) # bins is to resize the histogram
plt.show()
```

#### The histogram plot is a simple graph that can be plotted for individual columns. The color can be specified as per to the requirement. The following plot is a combined result of 3 columns that is, The scores in math, reading and writing. The other options of histogram include bins which helps in resizing the hsitogram. To increase the transparency of the plot the option as alpha is used and it has the values between 0-1

```python
stu.plot()
```

#### Here the dataframe object stu of the pandas library helps in the direct plot of the columns of the dataset. In this plot, Scatter kind is used by default. There are many points that are plotted to give the combined result. These are not line plot and are individual points for individual values resulting in this graph.

```python
plt.scatter(stu.math_score,stu.reading_score,stu.writing_score)
```

#### The scatter graph of all the values in marks are plotted. Each value is an individual point and is represented in the graph.

```python
sb.jointplot(stu.percentages,stu.total_scores,kind='scatter')
```

#### The joint plot is a plotting function of seaborn. It is a combination of 2 plots that is, Scatter and Histogram.

```python
sb.jointplot(stu.reading_score,stu.writing_score,kind='kde')
```

#### The kde stands for kernel density estimate. This graph is an estimation of the probability density function. Helps in estimating the distribution of the data set

```python
sb.kdeplot(stu.total_scores)
```

#### The kde plot of a graph is the distribution estimation graph and is plotted here as a scatter graph. This helps in plotting a scatter graph which includes of only points and are extended to a line plot.

```python
sb.distplot(stu.total_scores,bins=25)
```

#### The distribution plot is a combination of the histogram and the kde scatter plot. bins is used in resizing the histogram.

```python
sb.boxplot(data=[stu.math_score,stu.reading_score,stu.writing_score])
```

#### The boxplot is for plotting a range of values for the selected columns. This is to represent the second and third quartiles, usually with a vertical line inside to indicate the median value. The lower and upper quartiles are shown as horizontal lines either side of the rectangle.

```python
v=stu[["math_score","reading_score"]]
sb.kdeplot(v)
```

#### The kde plot can be used for an array of columns and this shows the joint density of both the columns in the plot. The center of the plot indicates the presence of the maximum range of values in the dataset while the outer lines indicate that the values are reduced.
